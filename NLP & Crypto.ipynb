```python
#api key
secret = 'b198ef97107247fd87d110e9af018212'
# Define the endpoint
url = 'https://newsapi.org/v2/everything?'
# Specify the query and
# number of returns
parameters = {
    'q': "Tether", # query phrase
    'sortBy':'relevancy',
    "pagesize": 100, # maximum is 100
    "apikey": secret # your own API key
}
# Make the request
response = requests.get(url, params = parameters)
#Convert the response to JSON format and pretty print it
response_json = response.json()
pprint.pprint(response_json)

tether_sentiments = []
tether_lsentiment = []
tether_sentence = []
tether_semantics = []
tether_moodtags = []
tether_sentics = []
for i in range(100):
    try:
        url = response_json['articles'][i]['url']
        article = Article(url, config=config)
        try:
            article.download()
            article.parse()
        except:
            continue
        article.nlp()
        print("Number" , i + 1)
        text = article.summary
        date = article.publish_date
        sentiment = analyzer.polarity_scores(text)
        compound = sentiment["compound"]
        pos = sentiment["pos"]
        neu = sentiment["neu"]
        neg = sentiment["neg"]
        tether_sentiments.append({"text": text, "date": date, "compound": compound, "positive": pos, "negative": neg, "neutral": neu})
           
        text = article.summary
        sp = SenticPhrase(text)
        sentiment = sp.info(text)
        tether_sentence.append(text)
        tether_sentics.append(sp.get_sentics())
        tether_moodtags.append(sp.get_moodtags())
        tether_semantics.append(sp.get_semantics())
        tether_lsentiment.append(sp.info(text))
        print(sp.info(text))
        print('\n')
    
    except AttributeError:
        pass
```

```python
#Polarity
polarity = 0
for i in range(len(tether_lsentiment)):
    polarity += tether_lsentiment[i]['polarity']
print("Mean polarity = ",polarity/(len(tether_lsentiment)))

#Moodtags
from collections import Counter
# Moodtags
tether_moodtagswords = []
for l in tether_moodtags:
    for word in l:
        tether_moodtagswords.append(word)

tether_frequencytag = {}
# iterating over the list
for item in tether_moodtagswords:
    # checking the element in dictionary
    if item in tether_frequencytag:
    # incrementing the counr
        tether_frequencytag[item] += 1
    else:
    # initializing the count
        tether_frequencytag[item] = 1

# printing the frequency
print(tether_frequencytag)

# Finding 3 highest values
hightag = Counter(tether_frequencytag).most_common(20)
print("Dictionary with 3 highest values:")
print("Keys: Values")
for i in hightag:
    print(i[0]," :",i[1]," ")
    
for i in hightag:
    plt.bar(i[0], i[1])
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.xticks(fontsize = 25,rotation = 45)
plt.yticks(fontsize = 25)
plt.xlabel("Moodtag",fontsize = 30)
plt.ylabel("Number of tags",fontsize = 30)
plt.title("Number of tags associated with each moodtag",fontsize = 50)
plt.show()
```
